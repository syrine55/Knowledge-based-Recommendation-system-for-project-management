```python
import pdfplumber
import pandas as pd
import re
import io
import string
from tqdm import tqdm
```

## ExportingTitle and corpus: 

<ul>
    <li>Remove page number</li>
    <li>remove ฀฀</li>
    <li>Remover page footer text</li>
    <li>The first three can be found in the first two lines</li>
    <li>Bullet points = uu</li>
    <li>Title, subtitles, subsubtitles, etc are all upper case</li>
</ul>

```python
def clean_text(text, index):
    text = text.replace(str(index-38),"")
    text = text.replace("฀", "")
    text = text.replace("Part 1 - Guide","")
    text = text.replace("Guide","")
    text = re.sub(r"uu","",text)
    liste = text.split("\n")
    return liste
```

```python

with pdfplumber.open("PMBOK_6thEd_English.pdf") as file:
    ##Extract scope
    corpus=""
    title = ""
    doc = {"Title":[],"Corpus":[],"Page":[]}
    for idx in tqdm(range(167,210)):
        page = file.pages[idx]
        text = page.extract_text()
        liste = clean_text(text, idx)
        j = 0
        # print(liste)
        while j < len(liste):
            
            if liste[j].isdecimal():
                if liste[j+1].isupper():
                    title = liste[j]+". "+liste[j+1]
                    if corpus != "":
                        doc["Corpus"].append(corpus)
                        doc["Page"].append(idx+1)
                        corpus=""
                    doc["Title"].append(title)
                j+=2
            else:
                if liste[j].isupper():
                    title = liste[j]
                    if corpus != "":
                        doc["Corpus"].append(corpus)
                        doc["Page"].append(idx+1)
                        corpus=""
                    if liste[j+1].isupper():
                        corpus+=liste[j+1]
                        
                    doc["Title"].append(title)
                else:
                    corpus+=" "+liste[j]
                    # print(corpus)
                j+=1
    doc["Corpus"].append(corpus)
    doc["Page"].append(idx+1)
df_scope = pd.DataFrame(doc)
```

```python
## initiating empty dict and dict od data frames
doc={
    "Title_1":[],
    "Title_2":[],
    "Title_3":[],
    "Title_4":[],
    "Corpus":[],
}

dict_dfs = {
    5:df_scope,
    6:df_schedule,
    7:df_cost,
}
```

```python
##Iterating by chapter section
for sec in tqdm(range(5,8)):# 5,6 and 7
    sec_str = str(sec)+"."
    # print("////////// ",sec," //////////////")
    doc["Title_1"].append(table.loc[table.Section.str.contains(r"^"+sec_str+"$"), "Domain"].values[0]) 
    doc["Title_2"].append("")
    doc["Title_3"].append("")
    doc["Title_4"].append("")
    doc["Corpus"].append(dict_dfs[sec].loc[dict_dfs[sec].Title.str.contains(r"^"+sec_str+" [a-zA-Z]"), "Corpus"].values[0])
    if sec != 7: 
        for s_sec in range(1,7):
            s_str = str(sec)+"."+str(s_sec)
            doc["Title_1"].append(table.loc[table.Section.str.contains(r"^"+sec_str+"$"), "Domain"].values[0]) 
            doc["Title_2"].append(table.loc[table.Section.str.contains(r"^"+s_str+"$"), "Process"].values[0])
            doc["Title_3"].append("")
            doc["Title_4"].append("")
            doc["Corpus"].append(dict_dfs[sec].loc[dict_dfs[sec].Title.str.contains(r"^"+s_str+" [a-zA-Z]"), "Corpus"].values[0])
            
            for ss_sec in  range(1,4):
                ss_str = str(sec)+"."+str(s_sec)+"."+str(ss_sec)
                nbre_sss = dict_dfs[sec].loc[dict_dfs[sec].Title.str.contains(r""+ss_str+".\d")].shape[0]
                ## Looping through susbsubsub sections
                for sss_sec in range(1,nbre_sss+1):
                    sss_str = str(sec)+"."+str(s_sec)+"."+str(ss_sec)+"."+str(sss_sec)
                    # print(sss_str)
                    doc["Title_1"].append(table.loc[table.Section.str.contains(r"^"+sec_str+"$"), "Domain"].values[0]) 
                    doc["Title_2"].append(table.loc[table.Section.str.contains(r"^"+s_str+"$"), "Process"].values[0])
                    doc["Title_3"].append(table.loc[table.Section.str.contains(r"^"+ss_str+"$"), "type"].values[0])
                    doc["Title_4"].append(dict_dfs[sec].loc[dict_dfs[sec].Title.str.contains(r""+sss_str+""), "Title"].values[0].replace(""+sss_str+" ","").capitalize())
                    doc["Corpus"].append(dict_dfs[sec].loc[dict_dfs[sec].Title.str.contains(r""+sss_str+" [a-zA-Z]"), "Corpus"].values[0])
                # print("---------------------------")
    else: ##chapter 7 has 4 sub sections
        for s_sec in range(1,5):
            s_str = str(sec)+"."+str(s_sec)
            doc["Title_1"].append(table.loc[table.Section.str.contains(r"^"+sec_str+"$"), "Domain"].values[0]) 
            doc["Title_2"].append(table.loc[table.Section.str.contains(r"^"+s_str+"$"), "Process"].values[0])
            doc["Title_3"].append("")
            doc["Title_4"].append("")
            doc["Corpus"].append(dict_dfs[sec].loc[dict_dfs[sec].Title.str.contains(r"^"+s_str+" [a-zA-Z]"), "Corpus"].values[0])
            
            for ss_sec in  range(1,4):
                ss_str = str(sec)+"."+str(s_sec)+"."+str(ss_sec)
                nbre_sss = dict_dfs[sec].loc[dict_dfs[sec].Title.str.contains(r""+ss_str+".\d")].shape[0]
                ## Looping through susbsubsub sections
                for sss_sec in range(1,nbre_sss+1):
                    sss_str = str(sec)+"."+str(s_sec)+"."+str(ss_sec)+"."+str(sss_sec)
                    # print(sss_str)
                    doc["Title_1"].append(table.loc[table.Section.str.contains(r"^"+sec_str+"$"), "Domain"].values[0]) 
                    doc["Title_2"].append(table.loc[table.Section.str.contains(r"^"+s_str+"$"), "Process"].values[0])
                    doc["Title_3"].append(table.loc[table.Section.str.contains(r"^"+ss_str+"$"), "type"].values[0])
                    doc["Title_4"].append(dict_dfs[sec].loc[dict_dfs[sec].Title.str.contains(r""+sss_str+""), "Title"].values[0].replace(""+sss_str+" ","").capitalize())
                    doc["Corpus"].append(dict_dfs[sec].loc[dict_dfs[sec].Title.str.contains(r""+sss_str+""), "Corpus"].values[0])
                    
```
