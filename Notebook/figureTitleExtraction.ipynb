```python
#import libraries
import pdfplumber
import pandas as pd
import re
import io
import string
from tqdm import tqdm
```

```python
#import the database
df = pd.read_json("DataBase.json")
df
```

```python
df_corpus = df["Corpus"]
df_corpus
```

```python
#search for figures
figure=""
doc = {"figure":[]}
for i in range(len(df_corpus)):
    result1 = re.findall(r"Figure [0-9]-[0-9]+", df["Corpus"][i])
    result = list(set(result1))
    print(result)
    doc["figure"].append(result)
    df_figure = pd.DataFrame(doc)
```

```python
#search for tables
tables=""
doc = {"tables":[]}
for i in range(len(df_corpus)):
    result = re.findall(r"Table [0-9]-[0-9]", df["Corpus"][i])
    result = list(set(result))
    print (result)
    doc["tables"].append(result)
    df_tables = pd.DataFrame(doc)
```

```python
df = pd.DataFrame(list(zip(df["Title_1"],df["Title_2"],df["Title_3"],df["Title_4"],df["Corpus"],df_figure["figure"],df_tables["tables"])),columns=['Title_1','Title_2','Title_3','Title_4','Corpus','Figure','Tables'])
df
```
